**Prepare Yaml File**

```python
!pip install pyyaml

import yaml

# Load the dataset
data = {
    'train': '/kaggle/input/vehicle-dataset-for-yolo/vehicle dataset/train',
    'val': '/kaggle/input/vehicle-dataset-for-yolo/vehicle dataset/valid',
    'nc': 6,
    'names': ['car', 'threewheel', 'bus', 'truck', 'motorbike', 'van']
}

# Yaml file path
yaml_file = '/kaggle/working/data.yaml'

# Save data to yaml file
with open(yaml_file, 'w') as f:
    yaml.dump(data, f)
```

**Load the YOLOv11 Model**

```python
from ultralytics import YOLO

# Load the model
model = YOLO("yolo11x.pt")
model.fuse()
```

```python
model.info()
```

**Train Yolov11 Model using Dataset**

```python
results = model.train(data="/kaggle/working/data.yaml", epochs=100, imgsz=640, device=[0, 1])
```
